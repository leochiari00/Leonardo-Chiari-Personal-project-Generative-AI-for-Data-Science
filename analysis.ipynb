# Avocado Price Analysis – Data Science Final Project

## About the Dataset

This dataset contains historical data on avocado prices and sales volume across multiple regions in the United States. The dataset is published by the Hass Avocado Board and includes details on different avocado types, sizes, and packaging.

**Dataset columns:**

- `Date`: Date of record
- `AveragePrice`: Average price of a single avocado
- `Total Volume`: Total amount of avocados sold
- `4046`: Volume of PLU 4046 (small Hass)
- `4225`: Volume of PLU 4225 (large Hass)
- `4770`: Volume of PLU 4770 (extra large Hass)
- `Total Bags`: Total number of avocado bags sold
- `Small Bags`: Small bag sales
- `Large Bags`: Large bag sales
- `XLarge Bags`: Extra large bag sales
- `type`: Avocado type – conventional or organic
- `year`: Year of the record
- `region`: Region or city where the record was collected

**Dataset source**: [https://www.kaggle.com/datasets/neuromusic/avocado-prices]


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('avocado.csv')

# Display initial rows
df.head()
```

```python
# Dataset info
df.info()

# Statistical summary
df.describe(include='all')
```

## 4. Data Cleaning and Preprocessing

Steps:
1. Convert `Date` to datetime
2. Handle missing values
3. One-hot encode categorical features
4. Drop unnecessary columns
5. Scale features

```python
# Copy dataset
data = df.copy()

# Convert 'Date'
data['Date'] = pd.to_datetime(data['Date'])

# Check for missing values
print(data.isnull().sum())
```

```python
# Encode categorical variables
data_encoded = pd.get_dummies(data, columns=['type', 'region'], drop_first=True)

# Drop 'Date' column
data_encoded = data_encoded.drop(['Date'], axis=1)

# Separate features and target
X = data_encoded.drop('AveragePrice', axis=1)
y = data_encoded['AveragePrice']

print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)
```

```python
# Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 6. Outlier Detection

Identify potential outliers for `AveragePrice` using IQR method.

```python
# Boxplot of 'AveragePrice'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['AveragePrice'])
plt.title('Boxplot of AveragePrice')
plt.show()
```

```python
# Compute IQR bounds
Q1 = df['AveragePrice'].quantile(0.25)
Q3 = df['AveragePrice'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['AveragePrice'] < lower) | (df['AveragePrice'] > upper)]
print(f"Number of outliers: {len(outliers)}")
```

### Decision:
- We may remove these outliers before modeling to improve performance.

```python
# Remove outliers
df_clean = df[(df['AveragePrice'] >= lower) & (df['AveragePrice'] <= upper)]
```

## 7. Model Training and Evaluation

We’ll implement:
- Linear Regression
- Polynomial Regression (degree 2)
- Ridge Regression (with hyperparameter tuning)

For each, we'll:
- Split into train/test
- Train model
- Evaluate RMSE, MAE, R²

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Use cleaned & encoded & scaled features
# If using df_clean: repeat preprocessing to get X_scaled_clean, y_clean
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("Linear Regression:")
print("RMSE:", mean_squared_error(y_test, y_pred_lr, squared=False))
print("MAE:", mean_absolute_error(y_test, y_pred_lr))
print("R²:", r2_score(y_test, y_pred_lr))
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

lr_poly = LinearRegression()
lr_poly.fit(X_train_poly, y_train)
y_pred_poly = lr_poly.predict(X_test_poly)

print("Polynomial Regression (degree=2):")
print("RMSE:", mean_squared_error(y_test, y_pred_poly, squared=False))
print("MAE:", mean_absolute_error(y_test, y_pred_poly))
print("R²:", r2_score(y_test, y_pred_poly))
```

```python
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha': [0.1, 1, 10, 100]}
grid = GridSearchCV(ridge, params, scoring='neg_root_mean_squared_error', cv=5)
grid.fit(X_train, y_train)

print("Best Ridge alpha:", grid.best_params_['alpha'])
y_pred_ridge = grid.predict(X_test)

print("Ridge Regression:")
print("RMSE:", mean_squared_error(y_test, y_pred_ridge, squared=False))
print("MAE:", mean_absolute_error(y_test, y_pred_ridge))
print("R²:", r2_score(y_test, y_pred_ridge))
```

## 8. Results Summary

| Model                 | RMSE       | MAE         | R²          |
|----------------------|------------|-------------|-------------|
| Linear Regression     | 0.2652     | 0.2003      | 0.5623      |
| Polynomial Regression | 14.9514    | 0.7855      | -1390.3266  |
| Ridge Regression      | 0.2652     | 0.2003      | 0.5624      |

---

**Best model:** Ridge Regression (alpha=10) performed slightly better than Linear Regression in terms of R² score. Both had similar RMSE and MAE values, but Ridge offers better regularization, making it more robust.

---

### Interpretation

- **Polynomial Regression** (degree=2) performed significantly worse than the other models. This is likely due to overfitting caused by adding polynomial complexity without regularization.
- **Linear Regression** and **Ridge Regression** both performed similarly, but Ridge Regression had a slightly higher R² score and provides regularization, making it more reliable in preventing overfitting.
- This analysis shows that while simple models can be effective, adding complexity (like polynomial terms) without proper control can drastically reduce performance. Regularization methods such as Ridge can help balance bias and variance.
